## 📴 1. Offline Mode — Smart Cache System

````python
import os, json, hashlib, pickle
from datetime import datetime, timedelta
from pathlib import Path

CACHE_DIR = Path('.lumindad_adperf_cache')
CACHE_DIR.mkdir(exist_ok=True)

class OfflineCache:
    """
    LumindAd Smart Cache — TTL-based offline persistence.
    Supports API responses, model artifacts, and processed data.
    Works seamlessly with 11-language global deployment.
    """
    def __init__(self, cache_dir: Path, ttl_hours: int = 24):
        self.cache_dir = cache_dir
        self.ttl = timedelta(hours=ttl_hours)

    def _key(self, identifier: str) -> Path:
        h = hashlib.md5(identifier.encode()).hexdigest()[:14]
        return self.cache_dir / f'{h}.pkl'

    def get(self, key: str):
        path = self._key(key)
        if path.exists():
            with open(path, 'rb') as f:
                entry = pickle.load(f)
            if datetime.now() - entry['ts'] < self.ttl:
                return entry['data']
            path.unlink()
        return None

    def set(self, key: str, data):
        path = self._key(key)
        with open(path, 'wb') as f:
            pickle.dump({'ts': datetime.now(), 'data': data}, f)

    def stats(self):
        files = list(self.cache_dir.glob('*.pkl'))
        size_kb = sum(f.stat().st_size for f in files) / 1024
        return {'entries': len(files), 'size_kb': round(size_kb, 2)}

cache = OfflineCache(CACHE_DIR, ttl_hours=6)
print(f'✅ LumindAd Offline Cache initialized')
print(f'📁 Cache dir: {CACHE_DIR.absolute()}')
print(f'⏱  TTL: 6 hours | Stats: {cache.stats()}')
````

## 📡 3. APIs en Tiempo Real con Fallback Global

````python
EXCHANGE_FALLBACK = {
    'rates': {
        'DOP':58.5,'MXN':17.2,'BRL':5.1,'ARS':900.0,'COP':3900.0,
        'EUR':0.93,'GBP':0.79,'JPY':149.5,'KRW':1330.0,'CNY':7.23,
        'AUD':1.53,'CAD':1.36,'CHF':0.90,'INR':83.1,'TRY':30.8,
        'RUB':91.0,'AED':3.67,'SAR':3.75,'ILS':3.72,'NGN':1550.0
    },
    'base': 'USD', 'source': 'offline_fallback'
}

def fetch_with_cache(url, key, fallback, timeout=8):
    cached = cache.get(key)
    if cached:
        print(f'  📁 {key}: cache hit')
        return cached, 'cache'
    try:
        r = requests.get(url, timeout=timeout)
        r.raise_for_status()
        data = r.json()
        cache.set(key, data)
        print(f'  🌐 {key}: live API')
        return data, 'api'
    except Exception as e:
        print(f'  📴 {key}: offline fallback ({str(e)[:45]})')
        cache.set(key, fallback)
        return fallback, 'offline'

print('📡 Loading global real-time data...\n')
exchange_data, src_ex = fetch_with_cache(
    'https://open.er-api.com/v6/latest/USD',
    'exchange_rates_adperf', EXCHANGE_FALLBACK)

weather_data, src_wt = fetch_with_cache(
    'https://api.open-meteo.com/v1/forecast?latitude=18.5&longitude=-69.9&current_weather=true',
    'weather_santo_domingo_adperf', {'temperature': 28.5, 'windspeed': 12.3})

rates = exchange_data.get('rates', EXCHANGE_FALLBACK['rates'])
print(f'\n💱 Exchange rates loaded: {len(rates)} currencies')
print(f'🌤  Weather (Santo Domingo): {weather_data.get("current_weather",weather_data).get("temperature",28.5)}°C')
print(f'✅ APIs: Exchange={src_ex} | Weather={src_wt}')
````

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import time

# ── Dark Enterprise Theme ──────────────────────────────────────────
plt.rcParams.update({
    'figure.facecolor':'#0f1117','axes.facecolor':'#1a1d27',
    'axes.edgecolor':'#2d3250','axes.labelcolor':'#e0e0e0',
    'xtick.color':'#e0e0e0','ytick.color':'#e0e0e0',
    'text.color':'#e0e0e0','grid.color':'#2d3250','grid.alpha':.35,
    'font.family':'DejaVu Sans','axes.titlesize':13,'axes.titleweight':'bold',
})
PALETTE = ['#7c3aed','#06b6d4','#10b981','#f59e0b','#ef4444',
           '#8b5cf6','#0891b2','#059669','#d97706','#dc2626',
           '#a78bfa','#67e8f9','#6ee7b7','#fcd34d','#fca5a5']

# ══════════════════════════════════════════════════════════════════════
# SYNTHETIC DATASET GENERATION
# 8,000 campaigns · 50+ markets · 5 platforms · realistic causality
# ══════════════════════════════════════════════════════════════════════
np.random.seed(42)
N = 8_000

platforms  = list(PLATFORM_PERF.keys())
p_weights  = [PLATFORM_PERF[p]['share'] for p in platforms]
objectives = ['conversions','traffic','awareness','retargeting','lead_gen']
bid_strats = ['target_roas','target_cpa','max_conversions','manual_cpc']
devices    = ['mobile','desktop','tablet']
ad_formats = ['image','video','carousel','story','text']

platform_col  = np.random.choice(platforms,  N, p=p_weights)
country_col   = np.random.choice(countries_list, N, p=weights)
objective_col = np.random.choice(objectives, N)
bid_col       = np.random.choice(bid_strats, N)
device_col    = np.random.choice(devices,    N, p=[0.65,0.28,0.07])
adformat_col  = np.random.choice(ad_formats, N, p=[0.30,0.35,0.20,0.10,0.05])

audience_size  = np.random.lognormal(mean=11.5, sigma=1.2, size=N).astype(int)
audience_size  = np.clip(audience_size, 500, 10_000_000)
creative_score = np.clip(np.random.normal(62, 18, N), 0, 100)
daily_budget   = np.round(np.random.lognormal(mean=4.8, sigma=1.4, size=N), 2)
daily_budget   = np.clip(daily_budget, 5, 50_000)
duration_days  = np.random.choice([7,14,21,30,45,60,90], N,
                                   p=[0.10,0.18,0.12,0.28,0.12,0.12,0.08])

# ── Compute performance metrics with realistic causality ─────────
def compute_ctr(plat, obj, fmt, score, device):
    base  = PLATFORM_PERF[plat]['ctr']
    obj_m = {'conversions':0.95,'traffic':1.10,'awareness':0.85,
              'retargeting':1.25,'lead_gen':0.90}
    fmt_m = {'video':1.35,'story':1.20,'carousel':1.10,'image':1.00,'text':0.70}
    dev_m = {'mobile':1.15,'desktop':0.95,'tablet':0.90}
    score_m = 0.70 + (score / 100) * 0.60
    ctr = base * obj_m[obj] * fmt_m[fmt] * dev_m[device] * score_m
    ctr += np.random.normal(0, ctr * 0.15)
    return max(0.1, round(ctr, 3))

def compute_cpc(plat, country, budget):
    base   = PLATFORM_PERF[plat]['cpc']
    meta   = COUNTRY_META.get(country, {'cpc_idx': 0.8})
    budget_m = 0.85 + min(np.log1p(budget / 500) * 0.15, 0.30)
    cpc = base * meta['cpc_idx'] * budget_m
    cpc += np.random.normal(0, cpc * 0.20)
    return max(0.05, round(cpc, 4))

def compute_roas(plat, country, obj, bid, score, ctr):
    base  = PLATFORM_PERF[plat]['roas']
    meta  = COUNTRY_META.get(country, {'roas_base': 2.5})
    obj_m = {'conversions':1.20,'traffic':0.80,'awareness':0.60,
              'retargeting':1.35,'lead_gen':1.10}
    bid_m = {'target_roas':1.15,'target_cpa':1.10,
              'max_conversions':1.05,'manual_cpc':0.90}
    score_m = 0.60 + (score / 100) * 0.80
    roas = (base * 0.4 + meta['roas_base'] * 0.6) * obj_m[obj] * bid_m[bid] * score_m
    avg_ctr = PLATFORM_PERF[plat]['ctr']
    if ctr > avg_ctr * 1.2: roas *= 1.08
    elif ctr < avg_ctr * 0.8: roas *= 0.92
    roas += np.random.normal(0, roas * 0.18)
    return max(0.3, round(roas, 3))

print('⚙️  Generating 8,000 ad campaigns...', end=' ')
t0 = time.time()
ctr_col  = np.array([compute_ctr(platform_col[i], objective_col[i],
                                  adformat_col[i], creative_score[i],
                                  device_col[i]) for i in range(N)])
cpc_col  = np.array([compute_cpc(platform_col[i], country_col[i],
                                  daily_budget[i]) for i in range(N)])
roas_col = np.array([compute_roas(platform_col[i], country_col[i],
                                   objective_col[i], bid_col[i],
                                   creative_score[i], ctr_col[i]) for i in range(N)])

impressions_col = np.round(daily_budget / (cpc_col * ctr_col / 100 + 1e-6)).astype(int)
impressions_col = np.clip(impressions_col, 100, 5_000_000)
clicks_col      = np.round(impressions_col * ctr_col / 100).astype(int)
spend_col       = np.round(daily_budget * duration_days, 2)
revenue_col     = np.round(spend_col * roas_col, 2)
region_col      = [COUNTRY_META.get(c, {'region':'NA'})['region'] for c in country_col]
lang_col        = [COUNTRY_META.get(c, {'lang':'en'})['lang'] for c in country_col]

df = pd.DataFrame({
    'platform':      platform_col,
    'country':       country_col,
    'region':        region_col,
    'lang':          lang_col,
    'objective':     objective_col,
    'bid_strategy':  bid_col,
    'device':        device_col,
    'ad_format':     adformat_col,
    'audience_size': audience_size,
    'creative_score':creative_score.round(1),
    'daily_budget':  daily_budget,
    'duration_days': duration_days,
    'impressions':   impressions_col,
    'clicks':        clicks_col,
    'ctr':           ctr_col,
    'cpc':           cpc_col,
    'total_spend':   spend_col,
    'revenue':       revenue_col,
    'roas':          roas_col,
    'high_performance': (roas_col >= 3.0).astype(int),  # TARGET
})
print(f'done in {time.time()-t0:.1f}s')
print(f'✅ Dataset: {df.shape[0]:,} campaigns × {df.shape[1]} columns')
print(f'   High-performance rate: {df["high_performance"].mean():.1%}')
print(f'   ROAS range: {df["roas"].min():.2f} – {df["roas"].max():.2f} (mean={df["roas"].mean():.2f})')
print(f'   Countries: {df["country"].nunique()} | Platforms: {df["platform"].nunique()}')
df.head(3)
````

## 🔠 6. Feature Engineering & Encoding

````python
df_encoded = df.copy()

# ── One-Hot Encoding for categorical features ─────────────────────
df_encoded = pd.get_dummies(
    df_encoded,
    columns=['platform','objective','bid_strategy','device','ad_format','region'],
    drop_first=False, dtype=int
)

# ── Country + lang dropped (high cardinality) → region captured by OHE ─
df_encoded = df_encoded.drop(columns=['country','lang'], errors='ignore')

# ── Log-transform skewed numerics ────────────────────────────────
for col in ['audience_size','daily_budget','impressions','clicks','total_spend','revenue']:
    if col in df_encoded.columns:
        df_encoded[f'{col}_log'] = np.log1p(df_encoded[col])
        df_encoded.drop(columns=[col], inplace=True)

# ── Derived interaction features ──────────────────────────────────
df_encoded['ctr_x_score']     = df['ctr'] * df['creative_score'] / 100
df_encoded['budget_per_day']  = df['daily_budget'] / df['duration_days']
df_encoded['efficiency']      = df['roas'] / (df['cpc'] + 1e-4)

# ── Remove target-leaking columns ────────────────────────────────
# roas, ctr, cpc, revenue, efficiency directly determine the target
leaky = ['roas','ctr','cpc','revenue_log','efficiency']
for c in leaky:
    if c in df_encoded.columns and c != 'high_performance':
        df_encoded.drop(columns=[c], inplace=True, errors='ignore')

FEATURE_NAMES = [c for c in df_encoded.columns if c != 'high_performance']

print(f'✅ {len(FEATURE_NAMES)} features after engineering & encoding')
print(f'   OHE: platform×5 + objective×5 + bid×4 + device×3 + format×5 + region×7')
print(f'   Derived: ctr_x_score · budget_per_day')
print(f'   Log-transformed: audience_size · daily_budget · impressions · clicks · total_spend')
print(f'\nFeature sample: {FEATURE_NAMES[:8]}')
````

## 📊 9. Análisis Dirigido — Creative Score × ROAS · Budget × Performance

> **Análisis central de las dos variables de mayor impacto** en la performance de
> campañas digitales: `creative_score` (calidad del creativo) y `daily_budget` (presupuesto).

| Variable | Hipótesis | Tipo |
|----------|-----------|------|
| `creative_score` | Mayor calidad creativa → mayor ROAS | Continua 0–100 |

## 🔥 11. Deep Learning Suite — MLP · LSTM · CNN-1D · Autoencoder

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers, callbacks, Model
from tensorflow.keras.layers import (
    Dense, Dropout, BatchNormalization, Input,
    LSTM, Conv1D, MaxPooling1D, GlobalAveragePooling1D,
    Reshape, Flatten
)

print(f'🔥 TensorFlow {tf.__version__} | GPU: {len(tf.config.list_physical_devices("GPU")) > 0}')
tf.random.set_seed(42)
np.random.seed(42)

N_FEATURES = X_train_sc.shape[1]
COMMON_CALLBACKS = [
    callbacks.EarlyStopping(
        monitor='val_auc', patience=12,
        restore_best_weights=True, mode='max', verbose=1),
    callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5, patience=6, min_lr=1e-7, verbose=1),
]
COMPILE_ARGS = dict(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy',
             keras.metrics.AUC(name='auc'),
             keras.metrics.Precision(name='precision'),
             keras.metrics.Recall(name='recall')]
)

# ─────────────────────────────────────────────────────────────────────
# MODEL A: MLP — Deep Feed-Forward Neural Network
# Architecture: Input → 256 → 128 → 64 → 32 → Output
# ─────────────────────────────────────────────────────────────────────
def build_mlp(n):
    inp = Input(shape=(n,), name='mlp_input')
    x = Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-3))(inp)
    x = BatchNormalization()(x); x = Dropout(0.3)(x)
    x = Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-3))(x)
    x = BatchNormalization()(x); x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x)
    x = BatchNormalization()(x); x = Dropout(0.2)(x)
    x = Dense(32, activation='relu')(x)
    out = Dense(1, activation='sigmoid', name='hp_prob')(x)
    return Model(inp, out, name='LumindAd_MLP')

# ─────────────────────────────────────────────────────────────────────
# MODEL B: LSTM — Sequential Campaign Pattern Learning
# Treats features as temporal sequence for pattern detection
# Architecture: Input → Reshape → LSTM(64) → LSTM(32) → Dense → Out
# ─────────────────────────────────────────────────────────────────────
def build_lstm(n):
    inp = Input(shape=(n,), name='lstm_input')
    x   = Reshape((n, 1))(inp)
    x   = LSTM(64, return_sequences=True, dropout=0.2, recurrent_dropout=0.1)(x)
    x   = LSTM(32, return_sequences=False, dropout=0.2)(x)
    x   = Dense(32, activation='relu')(x)
    x   = BatchNormalization()(x)
    x   = Dropout(0.25)(x)
    out = Dense(1, activation='sigmoid', name='hp_prob')(x)
    return Model(inp, out, name='LumindAd_LSTM')

# ─────────────────────────────────────────────────────────────────────
# MODEL C: CNN-1D — Local Feature Interaction Extraction
# Convolutional filters learn local correlations between ad features
# Architecture: Input → Reshape → Conv1D(64,3) → Conv1D(32,3) → GAP → Dense → Out
# ─────────────────────────────────────────────────────────────────────
def build_cnn1d(n):
    inp = Input(shape=(n,), name='cnn_input')
    x   = Reshape((n, 1))(inp)
    x   = Conv1D(64, kernel_size=3, activation='relu', padding='same')(x)
    x   = BatchNormalization()(x)
    x   = MaxPooling1D(pool_size=2)(x)
    x   = Conv1D(32, kernel_size=3, activation='relu', padding='same')(x)
    x   = BatchNormalization()(x)
    x   = GlobalAveragePooling1D()(x)
    x   = Dense(32, activation='relu')(x)
    x   = Dropout(0.25)(x)
    out = Dense(1, activation='sigmoid', name='hp_prob')(x)
    return Model(inp, out, name='LumindAd_CNN1D')

# ─────────────────────────────────────────────────────────────────────
# MODEL D: AUTOENCODER — Spend-Spike Anomaly Detection
# Learns to reconstruct normal campaigns; high error = abnormal campaign
# Architecture: Encoder(128→64→32→16) ↔ Decoder(16→32→64→128→Input)
# ─────────────────────────────────────────────────────────────────────
def build_autoencoder(n):
    inp        = Input(shape=(n,), name='ae_input')
    enc        = Dense(128, activation='relu')(inp)
    enc        = BatchNormalization()(enc); enc = Dropout(0.2)(enc)
    enc        = Dense(64, activation='relu')(enc)
    enc        = Dense(32, activation='relu')(enc)
    bottleneck = Dense(16, activation='relu', name='bottleneck')(enc)
    dec        = Dense(32, activation='relu')(bottleneck)
    dec        = Dense(64, activation='relu')(dec)
    dec        = BatchNormalization()(dec)
    dec        = Dense(128, activation='relu')(dec)
    out        = Dense(n, activation='linear', name='reconstruction')(dec)
    return Model(inp, out, name='LumindAd_Autoencoder')

mlp_model  = build_mlp(N_FEATURES)
lstm_model = build_lstm(N_FEATURES)
cnn_model  = build_cnn1d(N_FEATURES)
ae_model   = build_autoencoder(N_FEATURES)

for m in [mlp_model, lstm_model, cnn_model, ae_model]:
    params = m.count_params()
    print(f'📐 {m.name:<25} | params: {params:>8,}')
````

````python
# ── Train MLP ─────────────────────────────────────────────────────────
print('🧠 [1/4] Training MLP (Feed-Forward Neural Network)...')
mlp_model.compile(**COMPILE_ARGS)
t0 = time.time()
hist_mlp = mlp_model.fit(
    X_train_sc, y_train_sm,
    epochs=80, batch_size=256, validation_split=0.15,
    callbacks=COMMON_CALLBACKS, verbose=0)
mlp_time = time.time() - t0
print(f'  ✅ MLP done in {mlp_time:.1f}s | Epochs: {len(hist_mlp.history["loss"])}')

# ── Train LSTM ────────────────────────────────────────────────────────
print('🌊 [2/4] Training LSTM (Sequential Campaign Patterns)...')
lstm_model.compile(**COMPILE_ARGS)
t0 = time.time()
hist_lstm = lstm_model.fit(
    X_train_sc, y_train_sm,
    epochs=60, batch_size=256, validation_split=0.15,
    callbacks=COMMON_CALLBACKS, verbose=0)
lstm_time = time.time() - t0
print(f'  ✅ LSTM done in {lstm_time:.1f}s | Epochs: {len(hist_lstm.history["loss"])}')

# ── Train CNN-1D ──────────────────────────────────────────────────────
print('📡 [3/4] Training CNN-1D (Feature Interaction Extraction)...')
cnn_model.compile(**COMPILE_ARGS)
t0 = time.time()
hist_cnn = cnn_model.fit(
    X_train_sc, y_train_sm,
    epochs=60, batch_size=256, validation_split=0.15,
    callbacks=COMMON_CALLBACKS, verbose=0)
cnn_time = time.time() - t0
print(f'  ✅ CNN-1D done in {cnn_time:.1f}s | Epochs: {len(hist_cnn.history["loss"])}')

# ── Train Autoencoder (only on standard/low-performance campaigns) ─────
print('🔎 [4/4] Training Autoencoder (Anomaly Detection — standard campaigns)...')
ae_model.compile(optimizer=keras.optimizers.Adam(1e-3), loss='mse', metrics=['mae'])
# Train ONLY on standard campaigns (learns 'normal' ad pattern)
X_normal = X_train_sc[y_train_sm == 0]
t0 = time.time()
hist_ae = ae_model.fit(
    X_normal, X_normal,
    epochs=40, batch_size=256, validation_split=0.15,
    callbacks=[callbacks.EarlyStopping(monitor='val_loss', patience=8,
                                        restore_best_weights=True, verbose=1)],
    verbose=0)
ae_time = time.time() - t0
print(f'  ✅ Autoencoder done in {ae_time:.1f}s | Epochs: {len(hist_ae.history["loss"])}')
print('\n✅ All 4 deep learning models trained!')
````

````python
# ── Evaluate MLP, LSTM, CNN-1D ──────────────────────────────────────
for name, model, hist, t in [
    ('MLP (Neural Net)', mlp_model, hist_mlp, mlp_time),
    ('LSTM',            lstm_model, hist_lstm, lstm_time),
    ('CNN-1D',          cnn_model,  hist_cnn,  cnn_time),
]:
    y_prob = model.predict(X_test_sc, verbose=0).flatten()
    y_pred = (y_prob >= 0.5).astype(int)
    results[name] = {
        'accuracy':  accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall':    recall_score(y_test, y_pred),
        'f1':        f1_score(y_test, y_pred),
        'roc_auc':   roc_auc_score(y_test, y_prob),
        'cm': confusion_matrix(y_test, y_pred),
        'y_prob': y_prob, 'time': t,
        'color': {'MLP (Neural Net)':'#ef4444','LSTM':'#06b6d4','CNN-1D':'#10b981'}[name]
    }

# ── Evaluate Autoencoder as anomaly detector ─────────────────────────
recon_err = np.mean(np.square(X_test_sc - ae_model.predict(X_test_sc, verbose=0)), axis=1)
threshold = np.percentile(
    np.mean(np.square(X_normal - ae_model.predict(X_normal, verbose=0)), axis=1), 90)
ae_pred   = (recon_err > threshold).astype(int)
results['Autoencoder (AE)'] = {
    'accuracy':  accuracy_score(y_test, ae_pred),
    'precision': precision_score(y_test, ae_pred, zero_division=0),
    'recall':    recall_score(y_test, ae_pred),
    'f1':        f1_score(y_test, ae_pred),
    'roc_auc':   roc_auc_score(y_test, recon_err),
    'cm':   confusion_matrix(y_test, ae_pred),
    'y_prob': recon_err / recon_err.max(),
    'time': ae_time, 'color': '#a78bfa'
}

print('\n📊 FULL RESULTS — ALL 8 MODELS')
print('='*75)
print(f'{"Model":<22} {"Accuracy":>9} {"Precision":>10} {"Recall":>8} {"F1":>8} {"AUC":>8}')
print('-'*75)
for name, r in results.items():
    print(f'{name:<22} {r["accuracy"]:>9.4f} {r["precision"]:>10.4f} '
          f'{r["recall"]:>8.4f} {r["f1"]:>8.4f} {r["roc_auc"]:>8.4f}')
print('='*75)
best = max(results, key=lambda k: results[k]['f1'])
print(f'\n🏆 Best model: {best} — F1={results[best]["f1"]:.4f} | AUC={results[best]["roc_auc"]:.4f}')
````

## 🔬 14. Análisis de Overfitting / Underfitting — Diagnóstico de Generalización

| Problema | Definición | Señal | Solución Aplicada |
|----------|-----------|-------|------------------|
| **Overfitting** | Aprende demasiado del training set, no generaliza | Train acc >> Test acc | Dropout, L2, EarlyStopping, SMOTE |
| **Underfitting** | Modelo demasiado simple, no captura patrones | Baja accuracy en train Y test | Aumentar complejidad, más features |
| **Bien calibrado** | Generaliza correctamente | Train ≈ Test accuracy | Balance óptimo encontrado |

````python
# ══════════════════════════════════════════════════════════════════════
# ANÁLISIS OVERFITTING / UNDERFITTING — Todos los modelos
# Evalúa train vs test accuracy y diagnostica generalización
# ══════════════════════════════════════════════════════════════════════

overfit_data = {}
print('Calculando train vs test accuracy...\n')

# ── 1. Classical ML: calcular train accuracy ─────────────────────
for name, cfg in models_ml.items():
    Xtr = X_train_sc if cfg['scaled'] else X_train_sm.values
    Xte = X_test_sc  if cfg['scaled'] else X_test.values
    train_pred = cfg['model'].predict(Xtr)
    test_pred  = cfg['model'].predict(Xte)
    train_acc  = accuracy_score(y_train_sm, train_pred)
    test_acc   = accuracy_score(y_test, test_pred)
    gap        = train_acc - test_acc
    overfit_data[name] = {
        'train_acc': train_acc, 'test_acc': test_acc, 'gap': gap,
        'type': 'Classical ML',
        'diagnosis': 'Overfitting' if gap > 0.05 else ('Underfitting' if test_acc < 0.70 else 'Bien calibrado')
    }

# ── 2. Deep Learning: usar historial train vs val ─────────────────
for name, model, hist in [
    ('MLP (Neural Net)', mlp_model, hist_mlp),
    ('LSTM',            lstm_model, hist_lstm),
    ('CNN-1D',          cnn_model,  hist_cnn),
]:
    train_acc_f = hist.history['accuracy'][-1]
    val_acc_f   = hist.history['val_accuracy'][-1]
    gap = train_acc_f - val_acc_f
    overfit_data[name] = {
        'train_acc': train_acc_f, 'test_acc': results[name]['accuracy'], 'gap': gap,
        'type': 'Deep Learning', 'hist': hist,
        'diagnosis': 'Overfitting' if gap > 0.05 else ('Underfitting' if val_acc_f < 0.70 else 'Bien calibrado')
    }

# Autoencoder: MSE train vs val
ae_tl = hist_ae.history['loss'][-1]
ae_vl = hist_ae.history['val_loss'][-1]
ae_g  = ae_vl - ae_tl
overfit_data['Autoencoder (AE)'] = {
    'train_acc': 1 - ae_tl, 'test_acc': 1 - ae_vl, 'gap': ae_g,
    'type': 'Deep Learning',
    'diagnosis': 'Overfitting' if ae_g > 0.05 else 'Bien calibrado'
}

# ── 3. Tabla diagnóstico ─────────────────────────────────────────
DIAG_ICONS = {
    'Overfitting':    '🔴 Overfitting',
    'Underfitting':   '🟡 Underfitting',
    'Bien calibrado': '🟢 Bien calibrado'
}
print('='*78)
print('DIAGNÓSTICO: OVERFITTING / UNDERFITTING — Todos los Modelos')
print('='*78)
print(f'{"Modelo":<22} {"Tipo":<14} {"Train Acc":>10} {"Test Acc":>10} {"Gap":>8} {"Diagnóstico"}')
print('-'*78)
for name, d in overfit_data.items():
    icon = DIAG_ICONS[d['diagnosis']]
    print(f'{name:<22} {d["type"]:<14} '
          f'{d["train_acc"]:>10.4f} {d["test_acc"]:>10.4f} '
          f'{d["gap"]:>8.4f}  {icon}')
print('='*78)

# ── 4. Visualización ────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(22, 7))
fig.patch.set_facecolor('#0f1117')

names_ov  = list(overfit_data.keys())
train_acc = [overfit_data[n]['train_acc'] for n in names_ov]
test_acc  = [overfit_data[n]['test_acc']  for n in names_ov]
gaps      = [overfit_data[n]['gap']       for n in names_ov]
x_pos = np.arange(len(names_ov)); w = 0.38

bars_tr = axes[0].bar(x_pos - w/2, train_acc, w,
                       label='Train Accuracy', color='#7c3aed', alpha=0.85)
bars_te = axes[0].bar(x_pos + w/2, test_acc,  w,
                       label='Test Accuracy',  color='#06b6d4', alpha=0.85)
axes[0].axhline(0.80, color='white', linestyle='--', alpha=0.25, linewidth=1)
axes[0].set_xticks(x_pos)
axes[0].set_xticklabels([n[:12] for n in names_ov], rotation=45, ha='right', fontsize=8)
axes[0].set_ylim(0.5, 1.02); axes[0].set_ylabel('Accuracy')
axes[0].set_title('Train vs Test Accuracy — Todos los Modelos')
axes[0].legend(fontsize=9, framealpha=0.2); axes[0].grid(axis='y', alpha=0.1)

gap_colors = ['#ef4444' if g > 0.05 else ('#f59e0b' if g > 0.02 else '#10b981') for g in gaps]
bars_g = axes[1].barh(names_ov, gaps, color=gap_colors, edgecolor='white', linewidth=0.3)
for bar, val in zip(bars_g, gaps):
    axes[1].text(val + 0.001, bar.get_y() + bar.get_height()/2,
                  f'{val:+.4f}', va='center', fontsize=8)
axes[1].axvline(0.05, color='#ef4444', linestyle='--', alpha=0.5,
                 linewidth=1.5, label='Umbral overfitting (0.05)')
axes[1].axvline(0.00, color='white', linewidth=0.5, alpha=0.3)
axes[1].set_title('Gap (Train Acc − Test Acc)')
axes[1].set_xlabel('Gap'); axes[1].invert_yaxis(); axes[1].grid(axis='x', alpha=0.1)
from matplotlib.patches import Patch
axes[1].legend(handles=[
    Patch(facecolor='#ef4444', label='Overfitting (gap > 0.05)'),
    Patch(facecolor='#f59e0b', label='Leve (gap 0.02-0.05)'),
    Patch(facecolor='#10b981', label='Bien calibrado (gap < 0.02)'),
], fontsize=8, framealpha=0.2, loc='lower right')

dl_curves = [
    ('MLP',   hist_mlp, '#ef4444'),
    ('LSTM',  hist_lstm,'#06b6d4'),
    ('CNN-1D',hist_cnn, '#10b981'),
]
for name, hist, color in dl_curves:
    epochs = range(1, len(hist.history['loss'])+1)
    axes[2].plot(epochs, hist.history['loss'],     color=color, linewidth=2,
                  linestyle='-',  label=f'{name} Train')
    axes[2].plot(epochs, hist.history['val_loss'], color=color, linewidth=2,
                  linestyle='--', alpha=0.7, label=f'{name} Val')
axes[2].set_title('Learning Curves — DL (Train vs Val Loss)')
axes[2].set_xlabel('Época'); axes[2].set_ylabel('Loss (Binary Crossentropy)')
axes[2].legend(fontsize=7, framealpha=0.2, ncol=2); axes[2].grid(alpha=0.1)

plt.suptitle(
    'Diagnóstico Overfitting / Underfitting — 8 Modelos\n'
    'LumindAd · AdPerformance ML v1.0',
    fontsize=13, fontweight='bold', y=1.02, color='white')
plt.tight_layout()
plt.savefig('overfitting_underfitting_adperf.png', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()

# ── 5. Explicación detallada ─────────────────────────────────────
EXPLANATIONS = {
    'Logistic Regression': {
        'causa':   'Modelo lineal — captura relaciones lineales entre features de campaña',
        'solucion':'Regularización L2 (C=1.0) · RobustScaler aplicado'
    },
    'Random Forest': {
        'causa':   'Ensemble profundo — riesgo overfitting en datasets de campañas',
        'solucion':'max_depth=12, min_samples_leaf=5 · limita complejidad'
    },
    'XGBoost': {
        'causa':   'Boosting iterativo — sensible a ruido en métricas de ad spend',
        'solucion':'subsample=0.8, colsample_bytree=0.8 · regularización implícita'
    },
    'KNN': {
        'causa':   'Alta dimensionalidad (OHE) — curse of dimensionality',
        'solucion':'k=15 (vecinos grandes) · RobustScaler'
    },
    'MLP (Neural Net)': {
        'causa':   'Red profunda (256→128→64→32) — alto riesgo overfitting',
        'solucion':'BatchNorm + Dropout(0.3) + L2(0.001) + EarlyStopping(patience=12)'
    },
    'LSTM': {
        'causa':   'Memoria recurrente — puede memorizar secuencias del training set',
        'solucion':'dropout=0.2 + recurrent_dropout=0.1 + EarlyStopping'
    },
    'CNN-1D': {
        'causa':   'Filtros convolucionales — overfitting en features de campaña',
        'solucion':'GlobalAveragePooling1D + Dropout(0.25) + BatchNorm'
    },
    'Autoencoder (AE)': {
        'causa':   'Entrena solo en campañas estándar — riesgo memorizar clase mayoritaria',
        'solucion':'Bottleneck estrecho (16 neuronas) + EarlyStopping(patience=8)'
    },
}
print('\n' + '='*68)
print('ANÁLISIS DETALLADO — CAUSAS Y SOLUCIONES APLICADAS')
print('='*68)
for name, d in overfit_data.items():
    icon = DIAG_ICONS[d['diagnosis']]
    exp  = EXPLANATIONS.get(name, {})
    print(f'\n{icon} — {name}')
    print(f'  Train Acc: {d["train_acc"]:.4f} | Test Acc: {d["test_acc"]:.4f} | Gap: {d["gap"]:+.4f}')
    if exp:
        print(f'  Causa:    {exp["causa"]}')
        print(f'  Solución: {exp["solucion"]}')
print('='*68)
print('\nCONCLUSIÓN: Los modelos con Dropout, BatchNorm y EarlyStopping')
print('demostraron mejor generalización. SMOTE balanceó el dataset sin')
print('contaminar el test set. Pipeline logra modelos bien calibrados.')
````
